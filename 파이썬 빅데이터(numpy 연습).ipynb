```python
import numpy as np

count = int(input("몇장을 구매하시겠습니까 : "))

def make_lotto(count):
    lotto_num = []
    for i in range(count) :
        for j in range(6) :
            lotto_num = np.random.choice(range(1,46),6, replace = False)
            
        lotto_num.sort()
        print("{} 로또번호 : {}".format(i+1, lotto_num))  # {} <-> i+1 , {} <-> lotto_num , 리스트 출력시 format함수사용
make_lotto(count)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(-100, 100, 1000) # -100~100까지 랜덤한 수1000개 생성
y = np.random.randint(-100, 100, 1000)

mask1 = abs(x) > 50 # x에 저장된 값 중에서 50보다 큰 값
mask2 = abs(y) > 50 # y에 저장된 값 중에서 50보다 큰 값

# print(x[mask1])
# print(y[mask2])

x= x[mask1 + mask2]
y= y[mask1 + mask2]

plt.scatter(x, y, c = x, cmap = 'jet', alpha = 0.7) #산점도 출력(x,y가 50이상이므로 도넛모양이나옴)
plt.colorbar()
plt.show()
```

```python
import csv
import numpy as np
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)
next(data)

home = []
home2 = []
mn=1
result_name = ''
result =0
data = list(data)

name = input('지역의 이름을 입력해주세요 : ')
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype = int) / int(row[2]) 
        
for row in data :
    away = np.array(row[3:], dtype=int) / int(row[2])
    s = np.sum((home-away)**2)  # 제곱해서 부호 양수화
    
    if s < mn and name not in row[0] : # s가 최솟값mn보다 작다면... and  대연제1동이 아닌경우(자기자신 포함x)
        mn = s
        result_name = row[0]
        result = away


plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (10,5)) # 박스 사이즈 조절
plt.title(name + '지역과 가장 비슷한 인구 구조를 가진 지역')
plt.plot(home, label= name)
plt.plot(result, label= result_name)
plt.legend()
plt.show()
```
