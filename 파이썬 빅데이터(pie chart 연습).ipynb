## 파이차이 그리기

```python
import matplotlib.pyplot as plt
plt.pie([10,20])
plt.show()
```

```python
import matplotlib.pyplot as plt
size = [2441, 2312, 1031, 1233]
blood = ['A형','B형','AB형','O형']
color = ['darkmagenta', 'deeppink', 'hotpink', 'pink']
plt.axis('equal') # 그래프와 범례가 겹칠 경우 동등하게 비율 맞춰줌
plt.pie(size, labels= blood, autopct='%.1f%%', colors=color, explode=(0,0,0.1,0)) # atuopct : 비율을 퍼센트로 나타내줌 # explode : 간격띄우기(범위 0~1)
plt.legend()
plt.show()
```
