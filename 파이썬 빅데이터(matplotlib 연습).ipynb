```python
import matplotlib.pyplot as plt

plt.title('Corona Status by Season')  # 제목
season = ['spring', 'summer', 'autumn', 'winter'] # x축
corona = [500, 2000, 7000, 4000] # y축
plt.plot(season,corona) # 그래프그리기
plt.xlabel('season') # x축 이름
plt.ylabel('num of infectee') # y축 이름
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Corona Status by Season')  # 제목
season = ['spring', 'summer', 'autumn', 'winter'] # x축
corona = [500, 2000, 7000, 4000] # y축
plt.plot(season,corona) # 그래프그리기
plt.bar(season,corona, width = 0.5, color='gray')
plt.xlabel('season') # x축 이름
plt.ylabel('num of infectee') # y축 이름
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Corona Status by Season')  # 제목

season = ['spring', 'summer', 'autumn', 'winter'] # x축
corona = [500, 2000, 7000, 4000] # y축

plt.subplot(2,1,1) # 2행 1열의 1열에 그래프생성
plt.plot(season,corona) # 그래프그리기
 
plt.subplot(2,1,2) # 2행 1열의 2열에 그래프생성
plt.bar(season,corona, width = 0.5, color='gray')

plt.xlabel('season') # x축 이름
plt.ylabel('num of infectee') # y축 이름
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title("Desk's Annual rating")

dele = [7.35, 7.35, 7.09, 6.95, 6.89]
eriksen = [7.45, 7.53, 7.4, 7.08, 6.51]
son = [6.53, 7.14, 6.99, 7.1, 7.23]
kane = [7.53, 7.68, 7.6, 7.38, 7.2]
xs=['2015', '2016', '2017', '2018', '2019']

plt.plot(xs, dele, 'r-', label='Dele Alli') #빨간색의 실선타입
plt.plot(xs, eriksen, 'g--', label='Christian Eriksen')
plt.plot(xs, son, 'b-', label='Heung-Min Son')
plt.plot(xs, kane, 'k:', label='Harry Kane')

plt.legend()
plt.xlabel("Year")
plt.ylabel("Rating")
plt.show() # 별도의 창으로 보여줌 (쥬피터에서는 내장되어있어서 안써도 나오긴함)

```

```python
#산점도
import matplotlib.pyplot as plt

plt.title("Player Goal and Assists")

goals = [13, 9, 8, 4, 3]
assists = [2, 9, 4, 2, 0]
players = ["Kane", "Son", "Alli", "Moura", "Bergwiin"]
plt.scatter(goals, assists) # scatter() : 산점도로 표현

for goal, assist, player in zip(goals, assists, players) : # 한꺼번에 리스트데이터를 가져옴
    plt.annotate(player, xy = (goal, assist), xytext = (5,-5), textcoords = 'offset points') # annotate() : 주석달기

plt.xlim(0,15) # x범위 설정
plt.xlabel("Goals")
plt.ylabel("Assists")
plt.show()
```
