```python
# 최고기온 구하기
import csv
f = open('seoul.csv', 'r', encoding = 'cp949') # 읽기모드로 오픈
data = csv.reader(f, delimiter=',') # 콤마로 구분

next(data) # header부분만 따로 빼냄

max_temp = -999 # 초기화
max_date = '' # 날짜 초기화

for row in data :
    if row[4] == '' :  # 결측치 설정
        row[4] = -999
    row[-1] = float(row[-1]) # 끝에서 부터 첫번째 데이터 
    
    if max_temp < row[-1] : # 최고기온 비교
        max_temp = row[-1]
        max_date = row[0]
    
f.close()
print("기상 관측 아래 서울의 최고 기온 : ", max_temp, "도 입니다")
print("기상 관측 아래 서울의 최고 기온 날짜는 : ", max_date, "입니다")
```

```python
# 최저기온 구하기
import csv
f = open('seoul.csv', 'r', encoding = 'cp949') # 읽기모드로 오픈
data = csv.reader(f, delimiter=',') # 콤마로 구분

next(data) # header부분만 따로 빼냄

min_temp = 999 # 초기화
min_date = '' # 날짜 초기화

for row in data :
    if row[3] == '' :  # 결측치 설정
        row[3] = 990
    row[-2] = float(row[-2]) # 끝에서 부터 첫번째 데이터 
    
    if min_temp > row[-2] : # 최고기온 비교
        min_temp = row[-2]
        min_date = row[0]
    
f.close()
print("기상 관측 아래 서울의 최저 기온 : ", min_temp, "도 입니다")
print("기상 관측 아래 서울의 최저 기온 날짜는 : ", min_date, "입니다")
```

```python
# 내 생일의 기온 변화를 그래프로 그리기
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', 'r', encoding = 'cp949') # 읽기모드로 오픈
data = csv.reader(f, delimiter=',') # 콤마로 구분

next(data) 

result = []

for row in data :
    if row[4] != '' :
        result.append(float(row[4]))
        
plt.figure(figsize = (10,2)) # 인치단위로 출력 (x축:10인치, y축:2인치)
plt.plot(result, 'r')
plt.show()

f.close()
```

```python
# 내 생일의 기온 변화를 그래프로 그리기
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', 'r', encoding = 'cp949') # 읽기모드로 오픈
data = csv.reader(f, delimiter=',') # 콤마로 구분

next(data) 

result = []
low = []

for row in data :
    if row[4] != '' and row[3] != '' :
        if 1990 <= int(row[0].split('-')[0]) : # 1990년대 이후로만 저장
            if row[0].split('-')[1] == '07' and row[0].split('-')[2] == '05' :
                result.append(float(row[4])) #최고기온
                low.append(float(row[3])) #최저기온
        
# plt.figure(figsize = (10,2)) # 인치단위로 출력 (x축:10인치, y축:2인치)
plt.rc('font', family = 'Malgun Gothic') # 폰트설정
plt.title('내 생일의 기온 변화 그래프')
plt.plot(result, 'r', label = 'high')
plt.plot(low, 'b', label = 'low')
plt.legend()
plt.show()

f.close()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('seoul.csv', 'r', encoding = 'cp949') # 읽기모드로 오픈
data = csv.reader(f, delimiter=',') # 콤마로 구분

next(data) 

result = []
result_low = []
for row in data :
    month = row[0].split('-')[1]
    month_low = row[0].split('-')[1]
    if row[4]!='' and row[3]!='' :
        if month=='08' :
            result.append(float(row[4])) # 최고기온
        elif month_low =='01' :
            result_low.append(float(row[3])) # 최저기온

f.close()
plt.hist(result, bins=100, color='r')
plt.hist(result_low, bins=100, color='b')

plt.show()


```
