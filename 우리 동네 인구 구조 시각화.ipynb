## 우리 동네 인구 구조 시각화하기
※ 우리 동네 인구 구조 시각화 절차(알고리즘 설계)
1. 인구 데이터 파일을 읽어온다.
2. 전체 데이터에서 한 줄씩 반복해서 읽어온다.
3. 우리 동네에 대한 데이터인지 확인한다.
4. 우리 동네일 경우 0세부터 100세까지의 인구수를 순서대로 저장한다.
5. 저장된 연령별 인구수 데이터를 시각화한다.

```python
import csv
import matplotlib.pyplot as plt
f = open('age.csv')
data = csv.reader(f)

result = []

name =input('인구 구조를 알고 싶은 지역 입력 : ')

for row in data : 
    if name in row[0] : # 전체 row중 첫번째 필드만 탐색
        for num in row[3:] : 
            result.append(int(num)) # 데이터를 리스트화하기
print(result)
plt.plot(result)  # matplotlib를 이용하여 데이터 시각화하기
plt.show()
```

## 항아리 모양 그래프 그리기

```python
import csv

f= open('gender.csv')
data= csv.reader(f)

man = []
woman = []

for row in data :
    if '대연제3동' in row[0] :
        for i in range(101) :
            man.append(int(row[i+3])) # 3부터시작해서 103까지 100번 반복해서 읽어옴
            woman.append(int(row[-(i+1)])) # -1부터 -101까지 100번 반복해서 읽어옴
woman.reverse() # reverse() : 데이터를 역순으로 뒤집어줌
```

```python
import csv
import matplotlib.pyplot as plt

f= open('gender.csv')
data= csv.reader(f)

man = []
woman = []

for row in data :
    if '대연제3동' in row[0] :
        for i in row[3:104] :
            man.append(-int(i)) #음수로 저장
        for i in row[106:] :
            woman.append(int(i))
            
f.close()

# plt.rcParams['axes.unicode_minus'] = False # -가 안나올 때 사용하면됨
plt.rc('font', family = 'Malgun Gothic')  # 폰트설정
plt.title('대연제3동 지역의 남녀 성별 인구 분포') # 제목 설정
plt.barh(range(101), man , color='b', label='남성') # 0세부터 100세까지 남성의 비율을 시각화
plt.barh(range(101), woman , color='r', label='여성') # 그래프 수평으로 배치
plt.legend() #범례표시
plt.show()
            
```

```python
# 파이 그래프로 나타내려면 총인구수가 필요함
import csv
import matplotlib.pyplot as plt

f= open('gender.csv')
data= csv.reader(f)

man = 0
woman = 0

size = []

ingu = input("지역을입력하시오 : ")

for row in data:
    if ingu in row[0] :
        for i in range(101) :
            man = man + int(row[i+3])
            woman = woman + int(row[i+106])
        break  # 제주 통계 자료까지만 읽고 반복문 빠져나옴

size.append(man) # 남자 인구 합산
size.append(woman) # 여자 인구 합산
f.close()

color = ['skyblue', 'pink']
# plt.figure(figsize = (10,5))  # 픽셀 단위로 나타냄
plt.rc('font', family = 'Malgun Gothic')
plt.title(ingu +' 지역의 남녀 성별 인구 분포')
plt.axis('equal')
plt.pie(size, labels= ['남','여'], autopct='%.1f%%', colors=color, startangle = 90) # startangle = 90 : 반시계 만큼 90이동
plt.legend() #범례표시
plt.show()
            
```
